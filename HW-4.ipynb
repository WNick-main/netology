### Решение

````python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

def getOwner():
    '''Point 1.1
    Пользователь по команде "p" может узнать владельца документа по его номеру'''
    docId = input("Введите номер документа:")
    owner = list(rec['name'] for rec in documents if rec['number'] == docId)
    result = ''
    if owner:
        result = 'Владелец документа: ' + str(owner[0])
    return result

def getShelf():
    '''Point 1.2
    Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится'''
    docId = input("Введите номер документа:") 
    Shelf = list(Shelf for Shelf,docs in directories.items() if docId in docs)
    result = ''
    if Shelf:
        result = 'Документ хранится на полке: ' + str(Shelf[0])
    return result  

def printAddDocs():
    '''Point 1.3
    Пользователь по команде "l" может увидеть полную информацию по всем документам'''
    allDocs = []
    for doc in documents:
        Shelf = list(Shelf for Shelf,docs in directories.items() if doc['number'] in docs)[0]
        allDocs.append(f'№: {doc["number"]}, тип: {doc["type"]}, владелец: {doc["name"]}, полка хранения: {Shelf}')
    return '\n' + '\n'.join(allDocs)

def addShelf():
    '''Point 1.4
    Пользователь по команде "ads" может добавить новую полку'''   
    newShelf = input('Введите номер полки:')
    if directories.get(newShelf) is None:
        directories[newShelf] = []
        AcResult = 'Полка добавлена. '
    else:
        AcResult = 'Такая полка уже существует. '
    Result = AcResult + 'Текущий перечень полок: ' + ', '.join(set(directories.keys()))
    return Result

def delShelf():
    '''Point 1.5
    Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)'''   
    newShelf = input('Введите номер полки:')
    if directories.get(newShelf) is not None:
        if not directories[newShelf]:
            directories.pop(newShelf)
            AcResult = 'Полка удалена. '
        else:
            AcResult = 'На полке есть документа, удалите их перед удалением полки. '
    else:
        AcResult = 'Такой полки не существует. '
    Result = AcResult + 'Текущий перечень полок: ' + ', '.join(set(directories.keys()))
    return Result
#-------------------------------------------------------------------------------
def addDoc():
    '''Point 2.1
    Пользователь по команде "ad" может добавить новый документ в данные'''   
    
    number = input('Введите номер документа:')
    dicType = input('Введите тип документа:')
    name = input('Введите владельца документа:')
    Shelf = input('Введите полку для хранения:')

    if directories.get(Shelf) is not None:
        documents.append(dict(type= dicType, number=number, name=name))
        directories[Shelf].append(number)
        AcResult = 'Документ добавлен. '
    else:
        AcResult = 'Такой полки не существует. Добавьте полку командой as. \n'
    Result = AcResult + 'Текущий перечень полок: ' + ', '.join(set(directories.keys()))
    return Result

def delDoc():
    '''Point 2.2
    Пользователь по команде "d" может удалить документ из данных'''   
    
    docId = input("Введите номер документа:")
    docIndex = list(Index for Index, rec in enumerate(documents) if rec['number'] == docId)
    if docIndex:
        documents.pop(docIndex[0])
        Result = 'Документ удален. \n'
    else:
        Result = 'Документ не найден в базе.  \n'
    return Result

def moveDoc():
    '''Point 2.2
    Пользователь по команде "m" может переместить документ с полки на полку'''   
    
    docId = input("Введите номер документа:")
    newShelf = input('Введите номер полки:')
    curShelf = list(Shelf for Shelf,docs in directories.items() if docId in docs)
    docIndex = list(Index for Index, rec in enumerate(documents) if rec['number'] == docId)

    if len(docIndex) > 0:
        if len(curShelf) > 0:
            directories[newShelf[0]].append(docId)
            directories[curShelf[0]].remove(docId)
            Result = 'Документ перемещен. \n' + printAddDocs()
        else:
            Result = 'Такой полки не существует. Текущий перечень полок: ' + ', '.join(set(directories.keys()))
    else:
        Result = 'Документ не найден в базе.  \n' + printAddDocs()
    return Result

#-------------------------------------------------------------------------------

def menu(command):
    result = ''
    if command == 'p':
        result = getOwner()
    if command == 's':
        result = getShelf()
    if command == 'l':
        result = printAddDocs()
    if command == 'ads':
        result = addShelf()        
    if command == 'ds':
        result = delShelf()  
    #----------------------
    if command == 'ad':
        result = addDoc()  
        result = result + printAddDocs()
    if command == 'd':
        result = delDoc()  
        result = result + printAddDocs()
    if command == 'm':
        result = moveDoc()        
    #----------------------
    if command == 'q': 
        result = 'До свидания'
    if result:
        print(result)
    else:
        print('Документ не найден в базе')
    nextCommand = True
    if command == 'q': nextCommand = False
    return nextCommand


while menu(input("Введите команду:")):
    pass
````
